```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
churn = pd.read_csv("churn.csv")
churn
```

```python
churn.columns.values
```

```python
churn.describe()
```

```python
churn.dtypes
```

```python
churn.isnull().sum()
```

```python
# Data Manipulation: dropping columns that may not be neccesary. Also removed "Joined" in customer_customers as they offer no significant data.
churn = churn.drop(['Customer_ID', 'Latitude', 'Longitude'], axis = 1)
churn.drop(churn[churn['Customer_Status'] == 'Joined'].index, inplace=True)
churn
```

```python
gender = churn['Gender'].value_counts()
gender
```

```python
gender.plot(kind="pie", title='Gender Male/Female', autopct='%1.1f%%')
```

```python
churn_distribution = churn['Customer_Status'].value_counts()
churn_distribution
```

```python
churn_distribution.plot(kind="pie", title='churn distribution', autopct='%1.1f%%')
```

```python
gender_churn = churn.groupby(['Gender', 'Customer_Status']).size().unstack()
gender_churn
```

```python
gender_churn.plot(kind="bar", xlabel='Gender', ylabel='Number of Customer')
```

```python
churn_category = churn.groupby(["Churn_Category", 'Customer_Status']).size().unstack()
churn_category
```

```python
churn_category.plot.pie(subplots=True, figsize=(7, 7), autopct='%1.1f%%')
```

```python
churn_reason = churn.groupby(["Churn_Category", 'Churn_Reason']).size().sort_values(ascending=False)
churn_reason
```

```python
# Biggest reason for churn was Competitor had better devices. Here, we're using device protection plan as a mesure of good devices.
device_churn = churn.replace({'Device_Protection_Plan': {0.0: 'No', 1.0: 'Yes'}}).groupby(['Device_Protection_Plan', 'Customer_Status']).size().unstack().dropna()
device_churn
```

```python
device_churn.plot(kind="bar", xlabel='Device Protection Plan', ylabel='Number of Customer')
```

```python
# Second biggest reason for churn was competitor made better offer. Here we're looking foe the churn rate of different offers.
offer_churn = churn.groupby(['Offer', 'Customer_Status']).size().unstack().dropna()
offer_churn
```

```python
offer_churn.plot(kind="bar", xlabel='Offer Type', ylabel='Offer Type')
```

```python
# Third biggest reason for churn was Attitude of support person. Here we examine the churn rate of customers who has premium customer support. Customers who does not have premium support has a significanlty larger churn rate. It implies customers with premium support enjoys better support experinece. 
support_churn = churn.replace({'Premium_Tech_Support': {0.0: 'No', 1.0: 'Yes'}}).groupby(['Premium_Tech_Support', 'Customer_Status']).size().unstack().dropna()
support_churn
```

```python
support_churn.plot(kind="bar", xlabel='Premium Tech Support', ylabel='Number of Customer')
```

```python
data_churn = churn.replace({'Unlimited_Data': {0.0: 'No', 1.0: 'Yes'}}).groupby(['Unlimited_Data', 'Customer_Status']).size().unstack().dropna()
data_churn
```

```python
data_churn.plot(kind="bar", xlabel='Unlimited Data', ylabel='Number of Customer')
```

```python
# Attitude of service provider was the 7th biggest reason for churn. Our chart below shows customers who opt in for phone service also have a higher churn population. There needs to be improvement in the phone service to reduce churn and increase custometr satisfaction.
service_churn = churn.replace({'Phone_Service': {0.0: 'No', 1.0: 'Yes'}}).groupby(['Phone_Service', 'Customer_Status']).size().unstack().dropna()
service_churn
```

```python
service_churn.plot(kind="bar", xlabel='Phone Service', ylabel='Number of Customer')
```

```python
security_churn = churn.replace({'Online_Security': {0.0: 'No', 1.0: 'Yes'}}).groupby(['Online_Security', 'Customer_Status']).size().unstack().dropna()
security_churn
```

```python
security_churn.plot(kind="bar", xlabel='Online Security', ylabel='Number of Customer')
```

```python
city_counts = churn.groupby('City').size()
city_counts_filtered = city_counts[city_counts > 20]
city_counts_filtered
```

```python
city_churn = churn.groupby(['City', 'Customer_Status']).size().unstack().dropna()
city_result = city_counts_filtered.to_frame().join(city_churn)
city_result.drop(0, axis=1, inplace=True)
city_result = city_result.astype(int)
city_result
```

```python
city_result.plot(kind="bar", xlabel='Cities', ylabel='Number of Customer', figsize=(12, 10))
plt.legend(["Churned", "Stayed"]);
plt.show()
```

```python
internet_churn = churn.groupby(['Internet_Type', 'Customer_Status']).size().unstack()

internet_churn = internet_churn.assign(Total=internet_churn.sum(axis=1))
internet_churn
```

```python
internet_churn.plot(kind="bar", xlabel='Internet Type', ylabel='Number of Customer')
```
